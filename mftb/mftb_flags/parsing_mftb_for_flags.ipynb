```python
import pandas as pd
from tqdm import tqdm
import os
```

```python
rodeo = pd.read_csv('/home/marfa/Desktop/lab/mycofactocin/mftb/mftb_psi_rodeo/main_co_occur_rodeo_psi.csv')
rodeo = rodeo.loc[:,['Query', 'Nucleotide_acc']].drop_duplicates()
rodeo.index = rodeo['Nucleotide_acc']
del rodeo['Nucleotide_acc']
rodeo
```

```python
with open('acc_nucl_mftb_psi_rodeo_mmseq.txt') as file:
    content = file.readlines()
    content = [line.rstrip() for line in content]
```

```python
assembl = {k:'' for k in content}
for filename in os.listdir('/home/marfa/Downloads/ncbi-genomes-2022-10-21/'):
    if filename.endswith(".asm") or filename.endswith(".py"): 
        continue
    else:
        df = pd.read_csv("/home/marfa/Downloads/ncbi-genomes-2022-10-21/"+filename, sep="\t",comment = "#", names = ['Sequence-Name','Sequence-Role','Assigned-Molecule','Assigned-Molecule-Location/Type','GenBank-Accn','Relationship','RefSeq-Accn','Assembly-Unit','Sequence-Length','UCSC-style-name'])
        for nucl_acc in df['RefSeq-Accn']:
            if nucl_acc in content:
                assembl_id = filename.split('_')[:2]
                assembl_id = '_'.join(assembl_id)
                assembl[nucl_acc] = assembl_id
                continue
assembl = pd.DataFrame({'nucl_id': assembl.keys(), 'assembl_id': assembl.values()})
assembl = assembl.drop([670]) #пустая строка
assembl.index = assembl['nucl_id']
del assembl['nucl_id']
assembl
```

```python
assembl = assembl.join(rodeo)
assembl
```
